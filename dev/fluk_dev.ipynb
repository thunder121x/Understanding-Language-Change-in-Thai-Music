```python
from ytmusicapi import YTMusic

ytmusic = YTMusic()

def get_song_metadata(song_title: str, artist: str) -> dict:
    """Fetch Thai song metadata (album, year, artist) using YouTube Music API."""
    query = f"{song_title} {artist}"
    results = ytmusic.search(query, filter="songs")
    if not results:
        return {}

    # Choose the first relevant result
    song = results[0]

    # Try to get more complete info via videoId
    video_id = song.get("videoId")
    if video_id:
        details = ytmusic.get_song(video_id)
        year = None
        if "microformat" in details and "microformatDataRenderer" in details["microformat"]:
            pub_date = details["microformat"]["microformatDataRenderer"].get("uploadDate")
            if pub_date:
                year = pub_date.split("-")[0]
        elif details.get("videoDetails", {}).get("year"):
            year = details["videoDetails"]["year"]

        return {
            "title": song.get("title"),
            "album": song.get("album", {}).get("name"),
            "artist": song.get("artists", [{}])[0].get("name"),
            "year": year,
            "views": song.get("views"),
        }

    return {
        "title": song.get("title"),
        "album": song.get("album", {}).get("name"),
        "artist": song.get("artists", [{}])[0].get("name"),
        "year": song.get("year"),
        "views": song.get("views"),
    }
```

```python
info = get_song_metadata("ฝันลำเอียง", "แจ้")
print(info)
```
